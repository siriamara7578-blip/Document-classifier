```python
import pandas as pd

from sklearn.model_selection import train_test_split

from sklearn.feature_extraction.text import TfidfVectorizer

from sklearn.naive_bayes import MultinomialNB

from sklearn.pipeline import Pipeline

from sklearn.metrics import accuracy_score, classification_report
```

```python
data = {
    "text": [

        # SPORTS (10)

        "The football team won the championship",
        "India won the cricket world cup",
        "The tennis player won the match",
        "The coach trained the athletes",
        "The batsman scored a century",
        "The football league starts tomorrow",
        "The stadium was full of fans",
        "The player received the best athlete award",
        "The volleyball tournament begins next week",
        "The team celebrated their victory",

        # POLITICS (10)

        "The government passed a new law",
        "Election results were announced today",
        "The Prime Minister addressed the nation",
        "The parliament discussed the budget",
        "The minister held a press conference",
        "A new public policy was introduced",
        "Citizens voted in the election",
        "The president met foreign leaders",
        "The senate approved the bill",
        "The political debate lasted for hours",

        # TECHNOLOGY (10)

        "Artificial Intelligence is transforming industries",
        "Machine learning improves prediction accuracy",
        "Python is used for data science",
        "Cloud computing stores data online",
        "Cybersecurity protects computer systems",
        "Deep learning powers image recognition",
        "Software developers build mobile apps",
        "The latest smartphone uses AI",
        "Robots automate factory operations",
        "Big data helps businesses make decisions"

    ],

    "category": [

        "Sports","Sports","Sports","Sports","Sports",
        "Sports","Sports","Sports","Sports","Sports",

        "Politics","Politics","Politics","Politics","Politics",
        "Politics","Politics","Politics","Politics","Politics",

        "Technology","Technology","Technology","Technology","Technology",
        "Technology","Technology","Technology","Technology","Technology"
    ]
}

df = pd.DataFrame(data)

df.head()
```

```python
print(df.shape)

print(df["category"].value_counts())
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    df["text"],
    df["category"],
    test_size=0.2,
    random_state=42,
    stratify=df["category"]
)
```

```python
model = Pipeline([
    ("tfidf", TfidfVectorizer()),
    ("classifier", MultinomialNB())
])
```

```python
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)

print(predictions)
```

```python
accuracy = accuracy_score(y_test, predictions)

print("Accuracy:", accuracy)
```

```python
print(classification_report(y_test, predictions, zero_division=0))
```

```python
sample = ["The cricket team won the final match"]

prediction = model.predict(sample)

print("Predicted Category:", prediction[0])
```

```python
sample = ["The parliament passed a new education bill"]

prediction = model.predict(sample)

print("Predicted Category:", prediction[0])
```

```python
import joblib

joblib.dump(model, "document_classifier.pkl")

print("Model saved successfully!")
```
